# What is a groupby function and how to implement it?

```python
import pandas as pd
```

```python
data = pd.read_csv('Students_Result.csv')
data.head()
```

```python
data.Best_of_Assignments.value_counts()
```

```python
data
```

```python
data.groupby('Best_of_Assignments').count()
```

```python
data.groupby('Best_of_Assignments').mean()
```

```python
data[data['Best_of_Assignments'] == 7]
```

```python
(20+9+11)/3
```

```python
data.groupby('Best_of_Assignments').Quiz_2.mean()
```

```python
data[data.Best_of_Assignments == 7].Quiz_2.mean()
```

```python
data.groupby('Best_of_Assignments').Quiz_2.agg(['mean', 'count', 'min', 'max'])
```
